```python
def GetHuggy(api_token,reference):
  i = 0
  b = []
  api_token = api_token
  reference = reference
  headers = {
    'Content-Type': 'application/json',
    'Accept': 'application/json',
    'X-Authorization': f'Bearer {api_token}'}
  request = Request(f'https://api.huggy.app/v2/{reference}?page={i}', headers=headers)
  response_body = urlopen(request).read()
  while len(response_body) > 2:
    api_token = api_token
    request = Request(f'https://api.huggy.app/v2/{reference}?page={i}', headers=headers)
    response_body = urlopen(request).read()
    x = json.loads(response_body)
    i += 1
    b.append(x)
  return b
```
